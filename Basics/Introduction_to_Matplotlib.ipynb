# Q2 Implement data visualization techniques such as plotting, bar, histogram, pie etc. using matplotlib


```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(-10,10,0.5)
y1=x
y2=1/(1+np.exp(-x)) # sigmoid
y3=np.where(x>0,1,0) # threshold
y4=np.where(x>0,x,0) # ramp
y5=np.tanh(x)
############## ALL in one figure
plt.figure(figsize=(5,5))
# lw=line width
# c='color'
# ls='--','-.','-''
#marker='d','s','o','<','>','*'
plt.plot(x,y1,c='r',lw=2,ls='-.',marker='o',markerfacecolor='r'
         ,markeredgecolor='b',label='linear')
plt.plot(x,y2,label='sigmoid')
plt.plot(x,y3,label='threshold')
plt.plot(x,y4,label='ramp')
plt.plot(x,y5,label='tanh')
plt.legend()
plt.xlabel('X-->',c='b')
plt.ylabel('Y-->',c='r')
plt.title('Activation function',c='m')
plt.ylim([-2,2])
plt.xlim([-5,5])
#plt.axis('off')
plt.grid('on')
plt.text(1,1,'Hello')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(-10,10,0.5)
y1=x
y2=1/(1+np.exp(-x)) # sigmoid
y3=np.where(x>0,1,0) # threshold
y4=np.where(x>0,x,0) # ramp
y5=np.tanh(x)
############## ALL in one figure using subplot
plt.figure(figsize=(5,5))
plt.subplot(3,2,1)
plt.plot(x,y1,c='r',lw=2,ls='-.',marker='o',markerfacecolor='r'
         ,markeredgecolor='b',label='linear')
plt.subplot(3,2,2)
plt.plot(x,y2,label='sigmoid')
plt.subplot(3,2,3)
plt.plot(x,y3,label='threshold')
plt.subplot(3,2,4)
plt.plot(x,y4,label='ramp')
plt.subplot(3,2,5)
plt.plot(x,y5,label='tanh')
plt.show()
```

```python
############# Barg Graph
plt.figure(figsize=(5,10))
month=['JAN','FEB','MAR','APR','MAY','JUNE']
sal=[10,5,8,7,6,5]
plt.subplot(2,1,1)
plt.bar(month,sal,width=0.5,edgecolor='r',color='b')
for index, value in enumerate(sal):
    plt.text(index,value+.2,str(value))

plt.title('Bar Graph')
plt.xlabel('year')
plt.ylabel('Salary (in lakh)')
##################
plt.subplot(2,1,2)
plt.barh(month,sal,edgecolor='r',color='b')
for index, value in enumerate(sal):
    plt.text(value+.2, index,str(value))
plt.title('Horizontal Bar Graph')
plt.ylabel('month')
plt.xlabel('Salary (in lakh)')
```

```python
##### Histogram
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 500, density=True, facecolor='g', alpha=0.75)


plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.xlim(40, 160)
plt.ylim(0, 0.03)
plt.grid(True)
plt.show()
```

```python
########## pie chart
y = np.array([60, 20, 15, 5])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, explode = [0.2,0,0,0], shadow = True,autopct='%.2f%%')
#plt.legend(title='Four fruits:',loc=1)
plt.show() 
```

```python
######## Scatter plot
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y,c='r')
plt.scatter(x, y+3,c='b')
```

```python
########### Box plot
np.random.seed(10)
data_1 = np.random.normal(100, 10, 200)
data_2 = np.random.normal(90, 20, 200)
data_3 = np.random.normal(80, 30, 200)
data_4 = np.random.normal(70, 40, 200)
data = [data_1, data_2, data_3, data_4]
label=['data1','data2','data3','data4']
# Creating plot
bp = plt.boxplot(data,labels=label)
 
# show plot
plt.show()
```
