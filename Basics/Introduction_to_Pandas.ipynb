**mathematical operations**

```python
df.head()
```

```python
df['quantity'].sum()
```

```python
df['item_price'].sum() # Sum of item_price will not be printed as it is not a nuerical column
                       
```

```python
#.str.replace() - to replace a character value in a data column

df['item_price'] = df['item_price'].str.replace('$' , '')
```

```python
df.head()
```

```python
#astype() - to convert a value in a data column into numerical form

df['item_price'] = df['item_price'].astype(float)
df['item_price'].sum()
```

```python
df.dtypes
```

```python
print(df['item_price'].sum())
print(df['item_price'].mean())
print(df['item_price'].median())
```

**group by()**

```python
df.head()  # We can make groups based on several conditions
```

```python
#unique() - returns unique values in a data column
# 9 groups can be formed based on quantity as there are 9 unique values of quantity

df['quantity'].unique()
```

```python
#value_counts() - to get frequency of each unique values

df['quantity'].value_counts()
```

```python
df['item_name'].unique()
```

```python
len(df['item_name'].unique())  # unique() function returns the unique values in any column
```

```python
df['item_name'].value_counts()  # valus_counts() returns how many times that unique value came?
```

```python
df['item_price'].mean()  # we are taking item_price for overall data
```

```python
df.groupby('quantity')['item_price'].mean()  #we are taking mean for item_price for each quantity gp
```

```python
df.groupby('quantity')['item_price'].sum()   # we are taking sum of item_price for each quantity group
```

```python
df.groupby('quantity')['item_price'].agg(['mean','median','sum'])
```

```python
df.groupby('quantity')['item_price'].agg(['mean','median'])
```

```python
df.groupby('quantity')['item_price'].agg(['mean','median']).plot.bar()
```

```python
df[df['quantity'] == 3]
```

* mean amd median price value of each item orfered

```python
df.groupby('item_name')['item_price'].agg(['mean','median'])
```

```python
df.groupby('item_name')['item_price'].agg(['mean','median']).plot.barh(figsize=(12,10)) 
# use bar instead of barh for vertical graph. 
```

```python
df.groupby('item_name')['quantity','item_price'].sum()
```

```python
df.groupby('item_name')['quantity','item_price'].sum().sort_values('quantity')
```

**Titanic Passenger dataset**

```python
df = pd.read_csv('http://bit.ly/kaggletrain')
df.head()
```

* **Unique values and count of each unique values in - Survived, Pclass, Sex and Embarked**
* **Average, Min and Maximum price paid by customer travelling in each class of passenger**
* **Overall average age of passengers**
* **Average age of passenger travelling in each passenger class**
* **Average age of Male and Female passengers**

```python
#Unique values and count of each unique values in - Survived, Pclass, Sex and Embarked
```

```python
df['Survived'].value_counts()
```

```python
df['Pclass'].value_counts()
```

```python
df['Sex'].value_counts()
```

```python
df['Embarked'].value_counts()
```

```python
#Average, Min and Maximum price paid by customer travelling in each class of passenger
df.groupby('Pclass')['Fare'].agg(['min','max', 'mean','median','sum'])
```

```python
#Overall average age of passengers
df['Age'].mean()
```

```python
#Average age of passenger travelling in each passenger class
df.groupby('Pclass')['Age'].mean()
```

```python
#Average age of passenger travelling in each passenger class
df.groupby('Sex')['Age'].mean()
```
