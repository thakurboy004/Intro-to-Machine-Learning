### Q10 Implement clustering algorithm K-Means, K-Means++, Agglomerative and DBSCAN using scikitlearn.

```python
from sklearn.datasets import make_blobs
x,y=make_blobs(n_samples=150,n_features=2,centers=3,cluster_std=0.5,shuffle=True,random_state=0)
import matplotlib.pyplot as plt
plt.scatter(x[:,0],x[:,1],s=50,c='white',marker='o',edgecolor='black')
```

## Kmeans

```python
from sklearn.cluster import KMeans

dist=[]
for i in range(1,31):
    km=KMeans(n_clusters=i,init='random',n_init=10)
    km.fit(x)
    dist.append(km.inertia_)
plt.plot(range(1,31),dist,marker='o')
    
    
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=3,init='random',n_init=10)
y_km=km.fit_predict(x)
print(f'Distortion={km.inertia_}')
print('centroids')
print(km.cluster_centers_)
```

```python
plt.scatter(x[y_km==0,0],x[y_km==0,1],s=50,c='red',marker='s',edgecolor='black',label='CLUSTER1')
plt.scatter(x[y_km==1,0],x[y_km==1,1],s=50,c='green',marker='o',edgecolor='black',label='CLUSTER2')
plt.scatter(x[y_km==2,0],x[y_km==2,1],s=50,c='blue',marker='^',edgecolor='black',label='CLUSTER3')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],s=300,c='yellow',marker='*',edgecolor='black'
            ,label='Centroid')
plt.legend()
```

## Kmeans++

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=3,init='k-means++',n_init=10)
y_km=km.fit_predict(x)
print(f'Distortion={km.inertia_}')
print('centroids')
print(km.cluster_centers_)
```

```python
plt.scatter(x[y_km==0,0],x[y_km==0,1],s=50,c='red',marker='s',edgecolor='black',label='CLUSTER1')
plt.scatter(x[y_km==1,0],x[y_km==1,1],s=50,c='green',marker='o',edgecolor='black',label='CLUSTER2')
plt.scatter(x[y_km==2,0],x[y_km==2,1],s=50,c='blue',marker='^',edgecolor='black',label='CLUSTER3')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],s=300,c='yellow',marker='*',edgecolor='black'
            ,label='Centroid')
plt.legend()
```

## AgglomerativeClustering

```python
from sklearn.cluster import AgglomerativeClustering
km=AgglomerativeClustering(n_clusters=3)
y_km=km.fit_predict(x)
```

```python
plt.scatter(x[y_km==0,0],x[y_km==0,1],s=50,c='red',marker='s',edgecolor='black',label='CLUSTER1')
plt.scatter(x[y_km==1,0],x[y_km==1,1],s=50,c='green',marker='o',edgecolor='black',label='CLUSTER2')
plt.scatter(x[y_km==2,0],x[y_km==2,1],s=50,c='blue',marker='^',edgecolor='black',label='CLUSTER3')
plt.legend()
```

## DBScan

```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
x,y=make_moons(n_samples=200,noise=0.1)
plt.scatter(x[:,0],x[:,1])
```

```python
from sklearn.cluster import DBSCAN
db=DBSCAN(eps=0.2,min_samples=5,metric='euclidean')
y_db=db.fit_predict(x)
plt.scatter(x[y_db==0,0],x[y_db==0,1],s=50,c='red',marker='s',edgecolor='black',label='CLUSTER1')
plt.scatter(x[y_db==1,0],x[y_db==1,1],s=50,c='green',marker='o',edgecolor='black',label='CLUSTER2')
```
